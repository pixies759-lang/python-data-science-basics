```python
import pandas as pd
import numpy as np
```

```python
data = {
    "nombre": ["Ana", "Luis", "Maria", "Juan", "Ana", None],
    "edad": [23, 35, np.nan, 29, 23, 40],
    "ingresos": [15000, 30000, 20000, None, 15000, 50000],
    "ciudad": ["CDMX", "CDMX", "GDL", "MTY", "CDMX", None]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
```

```python
df["edad"].fillna(df["edad"].mean(),inplace=True),
df["ingresos"].fillna(df["ingresos"].median(),inplace=True),
df["ciudad"].fillna("desconocido",inplace=True)
```
