# Mini proyecto EDA – Ventas de una tienda

Objetivo: analizar las ventas de una tienda para identificar patrones,
productos más vendidos y comportamiento por ciudad.



```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = {
    "fecha": pd.to_datetime([
        "2024-01-05", "2024-01-06", "2024-01-07",
        "2024-01-08", "2024-01-09", "2024-01-10",
        "2024-01-11", "2024-01-12", "2024-01-13"
    ]),
    "producto": ["Laptop", "Mouse", "Teclado", "Laptop", "Monitor",
                 "Mouse", "Laptop", "Teclado", "Monitor"],
    "ciudad": ["CDMX", "CDMX", "GDL", "MTY", "CDMX",
               "GDL", "MTY", "CDMX", None],
    "ventas": [20000, 500, 800, 21000, 7000, 450, None, 850, 7200],
    "cantidad": [1, 2, 1, 1, 1, 3, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.describe()
```

```python

df["ventas"].fillna(df["ventas"].median(), inplace=True)
df["ciudad"].fillna("Desconocido", inplace=True)
```

```python
df.groupby("producto")["ventas"].sum().plot(kind="bar")
plt.title("Ventas totales por producto")
plt.xlabel("ventas")
plt.ylabel("productos")
plt.show
```

```python
df.groupby("ciudad")["ventas"].sum().plot(kind="bar")
plt.title("ventas por ciudad")
plt.xlabel("ciudad")
plt.ylabel("ventas")
plt.show()
```

## Conclusiones

- Laptops generan el mayor ingreso total.
- CDMX concentra la mayor parte de las ventas.
- Productos de bajo costo como mouse se venden más en cantidad pero generan menor ingreso.
