```python
def analizar_ventas(ventas):
    total = sum(ventas)
    promedio = total / len(ventas)
    return total, promedio

ventas_mensuales = [1200, 1350, 1100, 1600, 1500]
analizar_ventas(ventas_mensuales)
```
